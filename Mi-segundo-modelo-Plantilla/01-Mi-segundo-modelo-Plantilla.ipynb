```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, KBinsDiscretizer, OneHotEncoder, RobustScaler, Normalizer, QuantileTransformer, PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## <font color='264CC7'> Lectura de Datos </font>


Realiza la lectura de los datos y presenta las primeras filas del conjunto de datos.

```python
# Cargar el conjunto de datos
url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv'
df = pd.read_csv(url)
# Seleccionar las columnas de interés
columns = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','price']
df = df.loc[:, columns]
df.head(5)
```

---
## <font color='264CC7'> Características y etiquetas </font>

Separa las características y las etiquetas y mostrar la dimensión de cada elemento.

```python
# Dividir el conjunto de datos en características y etiquetas
features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors']
X = df.loc[:, features]
y = df.loc[:, ['price']]
```

---
## <font color='264CC7'> Conjunto de entrenamiento y prueba </font>

Divide los datos en dos conjuntos: uno para entrenamiento y otro para prueba y visualiza la dimensión de cada conjunto. Usar `random_state` igual al últmo número de tu cédula.

```python
# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Revisa la distribución de las etiquetas en los conjuntos de entrenamiento y prueba.

```python
print('Dimensiones de los conjuntos de datos:')
print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('X_test:', X_test.shape)
print('y_test:', y_test.shape)
```

---
## <font color='264CC7'> Entrenamiento </font>

Define un modelo de regresión lineal y entrenalo con los datos de entrenamiento.

```python
# Crear el modelo y ajustarlo
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Visualizar los coeficientes (en escala normalizada)
print('Ecuación del modelo:')
for feature, coef in zip(features, modelo.coef_[0]):
    print(f'{feature}: {coef}')

print('Intercepto:', modelo.intercept_[0])
```

Escribe un párrafo sobre qué hace cada escalado no visto en clase:

- **RobustScaler:** Utiliza la mediana y el rango intercuartílico (IQR) en lugar de la media y la desviación estándar. Es ideal cuando hay muchos outliers, ya que no se ve afectado por valores extremos.
- **Normalizer:** Normaliza cada FILA para que tenga norma 1. No escala columnas, sino vectores completos. Es útil en modelos basados en similitud como KNN o clustering sobre ángulos (cosine similarity).
- **QuantileTransformer:** Convierte la distribución original en una distribución uniforme o normal. Garantiza que todos los valores se repartan homogéneamente entre 0 y 1. Es útil cuando los datos no son gaussianos.
- **PowerTransformer:** Aplica transformaciones como Yeo-Johnson o Box-Cox para estabilizar la varianza y hacer la distribución más cercana a una normal. Ayuda cuando hay fuerte asimetría en los datos.
