```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, KBinsDiscretizer, OneHotEncoder, RobustScaler, Normalizer, QuantileTransformer, PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

---
## <font color='264CC7'> Lectura de Datos </font>


Realiza la lectura de los datos y presenta las primeras filas del conjunto de datos.

```python
# Cargar el conjunto de datos
url = 'https://raw.githubusercontent.com/mGalarnyk/Tutorial_Data/master/King_County/kingCountyHouseData.csv'
df = pd.read_csv(url)

# Seleccionar las columnas de interés
columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'price']
df = df.loc[:, columns]
df.head(5)
```

## <font color='264CC7'> Preprocesamiento de datos </font>

```python
# Normalización Min-Max
min_max_scaler = MinMaxScaler()
df_min_max_normalized = pd.DataFrame(min_max_scaler.fit_transform(df), columns=df.columns)
print("Normalización Min-Max:\n")
display(df_min_max_normalized.head())

# Estandarización
standard_scaler = StandardScaler()
df_standardized = pd.DataFrame(standard_scaler.fit_transform(df), columns=df.columns)
print("Estandarización:\n")
display(df_standardized.head())

# RobustScaler
robust_scaler = RobustScaler()
df_robust_scaled = pd.DataFrame(robust_scaler.fit_transform(df), columns=df.columns)
print("Escalado RobustScaler:\n")
display(df_robust_scaled.head())

# Normalizer
normalizer = Normalizer()
df_normalized = pd.DataFrame(normalizer.fit_transform(df), columns=df.columns)
print("Escalado Normalizer:\n")
display(df_normalized.head())

# QuantileTransformer
quantile_scaler = QuantileTransformer(output_distribution='uniform', random_state=0)
df_quantile_scaled = pd.DataFrame(quantile_scaler.fit_transform(df), columns=df.columns)
print("Escalado QuantileTransformer:\n")
display(df_quantile_scaled.head())

# PowerTransformer
power_scaler = PowerTransformer(method='yeo-johnson')
df_power_scaled = pd.DataFrame(power_scaler.fit_transform(df), columns=df.columns)
print("Escalado PowerTransformer:\n")
display(df_power_scaled.head())
```

Escribe un párrafo sobre qué hace cada escalado no visto en clase:

- **RobustScaler:** Utiliza la mediana y el rango intercuartílico (IQR) en lugar de la media y la desviación estándar. Es ideal cuando hay muchos outliers, ya que no se ve afectado por valores extremos.
- **Normalizer:** Normaliza cada FILA para que tenga norma 1. No escala columnas, sino vectores completos. Es útil en modelos basados en similitud como KNN o clustering sobre ángulos (cosine similarity).
- **QuantileTransformer:** Convierte la distribución original en una distribución uniforme o normal. Garantiza que todos los valores se repartan homogéneamente entre 0 y 1. Es útil cuando los datos no son gaussianos.
- **PowerTransformer:** Aplica transformaciones como Yeo-Johnson o Box-Cox para estabilizar la varianza y hacer la distribución más cercana a una normal. Ayuda cuando hay fuerte asimetría en los datos.
